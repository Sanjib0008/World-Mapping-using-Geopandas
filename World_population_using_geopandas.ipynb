```python
### 2nd World data
world=gpd.read_file(r'E:\Personal\Training Programme\Python\Datasets\World_Countries_with_population.shp')
```

```python
world.info()
```

```python
world.plot()
```

```python
import matplotlib.pyplot as plt

f, ax = plt.subplots(1, 1, figsize=(10, 10))
world.plot(ax=ax, column='CONTINENT', cmap='tab10',legend=True, legend_kwds={'loc': 'lower left'})
plt.title('World Map Colored by Continent')
plt.show()
```

```python
f, ax =plt.subplots(1,1,figsize=(10,10))
world.plot(ax=ax, column='POP_EST', cmap='RdYlGn_r', legend=True)
ax.set_title('High Density Populated Country')
```

```python
from matplotlib.colors import LogNorm

norm=LogNorm(vmin=world.POP_EST.min(),
             vmax=world.POP_EST.max())
f, ax =plt.subplots(1,1,figsize=(10,10))
world.plot(ax=ax, column='POP_EST', cmap='RdYlGn_r', edgecolor='w', linewidth=0.2,
            norm=norm)
ax.axis('off')
ax.set_title("Population Per Country", fontsize=14)
```
